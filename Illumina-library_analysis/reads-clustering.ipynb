```python
# ===== User-defined input/output =====
INPUT_FASTQ = "/Users/leandro/Desktop/github/NGS-data/ORF2library1_260411/raw-data/matched_reads_E3.fastq"
OUTPUT_FASTQ = "/Users/leandro/Desktop/github/NGS-data/ORF2library1_260411/raw-data/afterCMV-E3.fastq"
```

```python
#Extract 20bp following CMV primer sequence

PATTERN = "CTCCATATATGGGCTATGAACTAATGACCCC"
EXTRACT_LEN = 20


def process_fastq():
    with open(INPUT_FASTQ, "r") as infile, open(OUTPUT_FASTQ, "w") as outfile:
        while True:
            header = infile.readline()
            if not header:
                break  # End of file

            # Ensure this is a valid FASTQ header
            if not header.startswith("@"):
                continue

            sequence = infile.readline().strip()
            plus = infile.readline().strip()
            quality = infile.readline().strip()

            # Extract clean read name (remove '@' and anything after first space) 
            read_name = header[1:].strip().split()[0]

            # Search for the pattern ONLY in the sequence line
            idx = sequence.find(PATTERN)

            if idx != -1:
                start = idx + len(PATTERN)
                end = start + EXTRACT_LEN

                # Ensure enough bases exist after the pattern
                if end <= len(sequence):
                    extracted_seq = sequence[start:end]

                    # Write FASTA entry (2-line format)
                    outfile.write(f">{read_name}\n") 
                    outfile.write(f"{extracted_seq}\n")


if __name__ == "__main__":
    process_fastq()
```
